## Understand the data types of Dataset and downgrade its type to save memory


```python
for k in df.keys():
    if df[k].dtype != "object":
        print(k,df[k].dtype)
```

```python
for k in df.keys():
    if df[k].dtype != "object":
        print("Feature name: {}, dtype: {}, Min={}, Max={}".format(k,df[k].dtype,df[k].min(),df[k].max()))
```

```python
for k in df.keys():
    if "int" in str(df[k].dtype) and df[k].max() <= 255:
        df[k] = df[k].astype("int8")
    if "int" in str(df[k].dtype) and df[k].max() > 255:
        df[k] = df[k].astype("int16")
```

```python
df.dtypes
```

```python
for k in df.keys():
    if "float" in str(df[k].dtype) :
        df[k] = df[k].astype("float16")
```

## Fill missing values


```python
# Some important factors which helps to fill empty data

# if feature has categorical data then replace empty value using mode()

# if feature has discrete data and range is very less (1-10), go with mode(),because they will be categorical data else go with median()

# if feature has continous data then go with median()
```

```python
df[df=="?"]
```

```python
for k in df.keys():
    if "?" in df[k].unique():
        print(k,df[k].dtype)
        df[k] = df[k].replace({"?":np.nan})
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
df["num-of-doors"]
```

```python
df["num-of-doors"].mode()
```

```python
df["num-of-doors"] = df["num-of-doors"].fillna(df["num-of-doors"].mode()[0])
```

```python
df["normalized-losses"] = df["normalized-losses"].fillna(df["normalized-losses"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["bore"] = df["bore"].fillna(df["bore"].mode()[0])
```

```python
df["stroke"] = df["stroke"].fillna(df["stroke"].mode()[0])
```

```python
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].mode()[0])
```

```python
df["peak-rpm"] = df["peak-rpm"].fillna(df["peak-rpm"].mode()[0])
```

```python
df["price"] = df["price"].fillna(df["price"].mode()[0])
```
